```python
from utils import AiDevsTask, OpenAIChat, Downloader, ContextCompressor
import json
import os
from langchain.schema import (
    SystemMessage,
    HumanMessage,
    AIMessage
)

def check_file_size(file_name):
    file_size = os.path.getsize(file_name)
    print("File size:", file_size, "KB")
    return file_size

task_name = "optimaldb"
task = AiDevsTask()
task.get_token(task_name)
task_response = task.get_task_content()

task_msg = task_response['msg']
print("Task message: ", task_msg)
task_database = task_response['database']
print("Task database: ", task_database)
task_hint = task_response['hint']
print("Task hint: ", task_hint)

file_name = "3friends.json"
if not os.path.exists(file_name):
    downloader = Downloader()
    file_name = downloader.download_file(task_database)
    print("File downloaded:", file_name)

    local_directory = os.path.dirname(os.path.abspath(file_name))
    file_full_path = os.path.join(local_directory, file_name)

    compressor = ContextCompressor()
    compressed_file = {}
    max_atempts = 5
    attempts = 0

    while True:
        attempts += 1
        file_size = check_file_size(file_full_path)
        
        if attempts >= max_atempts or file_size <= 9:
            break
        
        print("Attempt:", attempts)
        new_file_path = os.path.join(local_directory, f"compressed_{attempts}.json")   
        compressed_file = compressor.Compress(file_full_path, new_file_path)
        file_full_path = new_file_path

    if compressed_file:
        context = ""
        for key, value in compressed_file.items():
            context += f"###{key}:\n-{'\n-'.join(value)}\n###\n"
        print(context)
        task.send_answer(context)
```
